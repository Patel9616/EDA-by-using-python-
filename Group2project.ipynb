```python
import matplotlib.pyplot as plt

regional_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(8, 6))
regional_sales.plot(kind='bar', color=['blue', 'orange', 'green', 'red'])
plt.ylabel('Total Sales (Millions)')
plt.title('Total Sales by Region')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt

# platform_sales was calculated in the previous turn
platform_sales = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
platform_sales.plot(kind='bar')
plt.xlabel('Platform')
plt.ylabel('Total Sales (Millions)')
plt.title('Total Global Sales by Platform')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

games_per_year = df.groupby('Year').size()

plt.figure(figsize=(12, 6))
games_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Games Sold')
plt.title('Number of Games Sold Per Year')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

publisher_region_sales = df.groupby('Publisher')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# To avoid clutter, let's focus on the top N publishers
top_n = 10  # You can adjust this number
top_publishers = publisher_region_sales.sum(axis=1).nlargest(top_n).index
publisher_region_sales_top = publisher_region_sales.loc[top_publishers]


plt.figure(figsize=(14, 7))
publisher_region_sales_top.plot(kind='bar', stacked=True, ax=plt.gca())
plt.xlabel('Publisher')
plt.ylabel('Total Sales (Millions)')
plt.title(f'Regional Sales Distribution for Top {top_n} Publishers')
plt.xticks(rotation=90)
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate regional sales by genre
genre_region_sales = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(10, 8))
sns.heatmap(genre_region_sales, annot=True, fmt=".1f", cmap="YlGnBu")
plt.xlabel('Region')
plt.ylabel('Genre')
plt.title('Regional Sales by Genre (Millions)')
plt.show()
```

```python
import matplotlib.pyplot as plt

top_10_games = df.nlargest(10, 'Global_Sales')

plt.figure(figsize=(12, 6))
plt.bar(top_10_games['Name'], top_10_games['Global_Sales'], color='skyblue')
plt.xlabel('Game Title')
plt.ylabel('Global Sales (Millions)')
plt.title('Top 10 Video Games by Global Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter for games with Global_Sales > 1 million
hit_games_df = df[df['Global_Sales'] > 1.0]

# Group by publisher and count the number of hit games
publisher_hit_games = hit_games_df['Publisher'].value_counts()

# Find the publisher with the most hit games
top_publisher_hit_games = publisher_hit_games.index[0]
max_hit_games_count = publisher_hit_games.iloc[0]

print(f"The publisher with the highest number of hit games (> 1 million global sales) is {top_publisher_hit_games} with {max_hit_games_count} hit games.")

# To avoid clutter, let's plot the top N publishers by hit games
top_n_publishers = 10 # You can adjust this number
publisher_hit_games_top_n = publisher_hit_games.nlargest(top_n_publishers)


plt.figure(figsize=(12, 6))
publisher_hit_games_top_n.plot(kind='bar', color='skyblue')
plt.xlabel('Publisher')
plt.ylabel('Number of Hit Games (> 1M Sales)')
plt.title(f'Number of Hit Games (> 1M Sales) per Publisher (Top {top_n_publishers})')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

regional_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
total_global_sales = regional_sales.sum()

# Calculate the percentage of global sales for each region
regional_sales_percentage = (regional_sales / total_global_sales) * 100

plt.figure(figsize=(8, 8))
plt.pie(regional_sales_percentage, labels=regional_sales_percentage.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Proportion of Global Sales by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data again
df = pd.read_csv('/content/Video Games Sales Data.csv', encoding='latin-1')
# Drop null values
df.dropna(inplace=True)

# Group by year and sum the global sales
yearly_global_sales = df.groupby('Year')['Global_Sales'].sum()

plt.figure(figsize=(12, 6))
yearly_global_sales.plot(kind='area')
plt.xlabel('Year')
plt.ylabel('Total Global Sales (Millions)')
plt.title('Total Global Video Game Sales Over Time')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(df['Year'], df['Global_Sales'], alpha=0.5) # Using alpha to handle overplotting
plt.xlabel('Release Year')
plt.ylabel('Global Sales (Millions)')
plt.title('Global Sales vs. Release Year')
plt.grid(True)
plt.show()
```
